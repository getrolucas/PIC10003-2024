## Análise Exploratória dos Dados

```python
import os
os.chdir('..')

import pandas as pd
import numpy as np
from utilsforecast.preprocessing import fill_gaps
from utilsforecast.plotting import plot_series
import matplotlib.pyplot as plt
from main.utils import ts_train_test_split

pd.set_option('display.float_format', lambda x: '%.2f' % x)

plt.style.use(['ggplot'])

%load_ext autoreload
%autoreload 2
```

```python
# leitura dos dados brutos
df = pd.read_csv(
    "nbs/data/data.csv",
    sep=';', 
    parse_dates=['ds'],
    dayfirst=True,
    dtype={'y': np.int32})
```

```python
# preenchendo os dados com os dias ausentes
df = fill_gaps(
    df=df,
    freq='D',
    start='per_serie',
    end='per_serie'
).fillna(0)
```

```python
# estatísticas descritivas
df.groupby("unique_id")['y'].describe()
```

```python
# plot das séries
plot_series(
    df=df, 
    max_ids=len(df.unique_id.unique()), 
    palette='tab20c'
)
```

```python
# dividindo dados em treino e teste
train_df, test_df = ts_train_test_split(df=df, train_size=0.7, id_col='unique_id')
```

```python
# salvando dados
train_df.to_csv("nbs/data/train_df.csv", index=False)
test_df.to_csv("nbs/data/test_df.csv", index=False)
```
